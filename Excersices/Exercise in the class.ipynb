```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get('https://www.imdb.com/chart/toptv/?ref_=nv_tvv_250')
```

```python
titleTag = soup.select('table')[0].find_all('a')
for tag in titleTag:
    print (tag.contents[0])
```

 ***helpful source to find child elements with class*** -> [This link](https://stackoverflow.com/questions/13202087/beautiful-soup-find-children-for-particular-div)

```python
spanTag = soup.find_all("span", {"class": "secondaryInfo"})

for tag in spanTag:
    print(tag.contents[0])
```
